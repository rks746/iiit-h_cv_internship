```python
from ultralytics import YOLO # framework for object detection, segmentation and classification 
from IPython.display import Image, display # helps display images directly 
from pathlib import Path
import urllib.request # Python module which fetches data from the web 
```

```python
# downloading first image
url = "https://ultralytics.com/images/bus.jpg"
urllib.request.urlretrieve(url, "bus.jpg")
print("Downloaded bus.jpg")
```

```python
print("\nRunning Object Detection...")
detect_model = YOLO("yolov8n.pt")     # detection model
detect_results = detect_model.predict(source="bus.jpg", save=True)
detect_output_path = Path(detect_results[0].save_dir) / "bus.jpg" # saves image 
display(Image(filename=str(detect_output_path))) # displays image
```

```python
print("\nRunning Object Segmentation...")
seg_model = YOLO("yolov8n-seg.pt")    # segmentation model
seg_results = seg_model.predict(source="bus.jpg", save=True)
seg_output_path = Path(seg_results[0].save_dir) / "bus.jpg"
display(Image(filename=str(seg_output_path)))
```
